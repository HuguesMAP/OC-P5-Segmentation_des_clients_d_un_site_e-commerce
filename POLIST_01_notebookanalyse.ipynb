```python
# %%flake8

""" create a function to display a bar plot of a column from a dataframe """


def bar_df_column(df, column, xlabel, ylabel, title, order=True, rotation=0):
    plt.figure(figsize=(15, 7))
    if order:
        sns.countplot(x=column, data=df,
                      order=df[column].value_counts().iloc[:].index)
    else:
        sns.countplot(x=column, data=df)
    plt.xticks(rotation=rotation)

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    # to save graph
    plt.savefig(title + '.png')

    plt.show()
```

```python
# %%flake8

""" function to compare date from different columns
    and return df of of date from col1 lower than the date of col2"""


def compare_date(df, col1, col2):
    df = df[(df[col1]-df[col2]).dt.days < 0]
    return df
```

```python
# %%flake8

# Check the logic date,
# order_purchase_timestamp < order_approved_at < order_delivered_carrier_date
# < order_delivered_customer_date
# (use function compare_date)

list_of_date_growing = ['order_purchase_timestamp', 'order_approved_at',
                        'order_delivered_carrier_date',
                        'order_delivered_customer_date']

k = 0
for i in list_of_date_growing:
    for j in list_of_date_growing[:k:-1]:
        df_orders_check = compare_date(df_orders_copy, j, i)
        df_orders_copy = df_orders_copy[~(
            df_orders_copy['order_id'].isin(df_orders_check['order_id']))]
    k += 1
print('Nb of orders : ', df_orders_copy.shape[0])
```

```python
# %%flake8

"""create function to display median and mean fo datas"""


def display_median_mean(datas):
    print('Median: {:0.2f}'.format(datas.median()))
    print('Mean: {:0.2f}'.format(datas.mean()))
```

```python
# %%flake8

"""create function to display histogram of a dataframe column"""


def hist_df_column(df, column, x_label, y_label, title, log=True):
    plt.style.use('seaborn-whitegrid')
    plt.figure(figsize=(15, 5))
    if log:
        sns.histplot(data=df, x=np.log(df[column]+1))
#     plt.axvline(df[column].median(), color='r',
#                 linestyle='dashed', linewidth=2, label='median')

#     plt.axvline(df[column].mean(), color='g',
#                 linestyle='dashed', linewidth=2, label='mean')

        plt.xlabel(x_label + ' (log)', fontsize=12)
        plt.title(title + ' (log)', fontsize=15)
    else:
        sns.histplot(data=df, x=df[column])
        plt.xlabel(x_label, fontsize=12)
        plt.title(title, fontsize=15)
        
    plt.ylabel(y_label, fontsize=12)

    # to save graph
    plt.savefig(title + '.png')

    plt.show()

    # use function display_median_mean to display median and mean
    display_median_mean(df[column])
```

```python
# %%flake8

# hist about Late_delivery_(hours) (use function hist_df_column)
hist_df_column(df_orders_analyse, 'Late_delivery_(hours)',
               'Nb of hours', 'Nb of order', 'Late delivery (hours) per order')
```

```python
# %%flake8

""" create a function to display a pie of a column from a dataframe """


def pie_df_column(datas, title, rotatelabels=0):
    plt.figure(figsize=(7, 7))
    plt.pie(datas.value_counts(), labels=datas.dropna().unique(),
            autopct='%1.1f%%', startangle=90, rotatelabels=rotatelabels)
    plt.title(title)
    
    # to save graph
    plt.savefig(title + '.png')
    
    plt.show()
```

```python
# %%flake8

# pie about the order deliveried in time or not (use function pie_df_column)
pie_df_column(
    df_orders_analyse['Late_delivery_(hours)'] == 0, 'Order delivery in time')
```
